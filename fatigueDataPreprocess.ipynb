```python
import xlrd
import os
import shutil
import time

classDataPath = 'D:\\trim_video\\AItest\\classofdata'

basePath = 'D:\\trim_video\\AItest\\data'
labelFolderName = 'label'
frameFolderName = 'frame'

labelFolderPath = os.path.join(basePath, labelFolderName)
frameFolderPath = os.path.join(basePath, frameFolderName)

startTime = time.time()
for header, labelFile in enumerate(os.listdir(labelFolderPath)):
    print(labelFile, ' is preprocessing...')
    
    labelFilePath = os.path.join(labelFolderPath, labelFile)
    labelData = xlrd.open_workbook(labelFilePath)
    print('sheets: ', str(labelData.sheet_names()))
    
    table = labelData.sheet_by_name('Sheet1')
    
    label = table.col_values(3)
    startFrameNum = table.col_values(4)
    endFrameNum = table.col_values(5)
    
    for i, fatigueDegree in enumerate(label):
        if fatigueDegree == 1:
            FDfolder = os.path.join(classDataPath, str(1))
            for imgNum in range(int(startFrameNum[i]), int(endFrameNum[i])):
                imgName = '{:0>10}'.format(imgNum+1) + '.jpg'
                imgPath = os.path.join(frameFolderPath, 'trim_') + labelFile.split('.')[0]
                outImgName = str(header) + imgName
                outPath = os.path.join(FDfolder, outImgName)
                shutil.copy(os.path.join(imgPath, imgName), outPath)
                
        elif fatigueDegree == 2:
            FDfolder = os.path.join(classDataPath, str(2))
            for imgNum in range(int(startFrameNum[i]), int(endFrameNum[i])):
                imgName = '{:0>10}'.format(imgNum+1) + '.jpg'
                imgPath = os.path.join(frameFolderPath, 'trim_') + labelFile.split('.')[0]
                outImgName = str(header) + imgName
                outPath = os.path.join(FDfolder, outImgName)
                shutil.copy(os.path.join(imgPath, imgName), outPath)
        
        elif fatigueDegree == 3:
            FDfolder = os.path.join(classDataPath, str(3))
            for imgNum in range(int(startFrameNum[i]), int(endFrameNum[i])):
                imgName = '{:0>10}'.format(imgNum+1) + '.jpg'
                imgPath = os.path.join(frameFolderPath, 'trim_') + labelFile.split('.')[0]
                outImgName = str(header) + imgName
                outPath = os.path.join(FDfolder, outImgName)
                shutil.copy(os.path.join(imgPath, imgName), outPath)
                
        elif fatigueDegree == 4:
            FDfolder = os.path.join(classDataPath, str(4))
            for imgNum in range(int(startFrameNum[i]), int(endFrameNum[i])):
                imgName = '{:0>10}'.format(imgNum+1) + '.jpg'
                imgPath = os.path.join(frameFolderPath, 'trim_') + labelFile.split('.')[0]
                outImgName = str(header) + imgName
                outPath = os.path.join(FDfolder, outImgName)
                shutil.copy(os.path.join(imgPath, imgName), outPath)
                
        elif fatigueDegree == 5:
            FDfolder = os.path.join(classDataPath, str(5))
            for imgNum in range(int(startFrameNum[i]), int(endFrameNum[i])):
                imgName = '{:0>10}'.format(imgNum+1) + '.jpg'
                imgPath = os.path.join(frameFolderPath, 'trim_') + labelFile.split('.')[0]
                outImgName = str(header) + imgName
                outPath = os.path.join(FDfolder, outImgName)
                shutil.copy(os.path.join(imgPath, imgName), outPath)
                
        elif fatigueDegree == 6:
            FDfolder = os.path.join(classDataPath, str(6))
            for imgNum in range(int(startFrameNum[i]), int(endFrameNum[i])):
                imgName = '{:0>10}'.format(imgNum+1) + '.jpg'
                imgPath = os.path.join(frameFolderPath, 'trim_') + labelFile.split('.')[0]
                outImgName = str(header) + imgName
                outPath = os.path.join(FDfolder, outImgName)
                shutil.copy(os.path.join(imgPath, imgName), outPath)
                
                
endTime = time.time()

print(endTime - startTime)
```
