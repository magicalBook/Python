```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

pts = np.random.rand(30)*.2
# Now let's make two outlier points which are far away from everything.
pts[[3, 14]] += .8

# If we were to simply plot pts, we'd lose most of the interesting
# details due to the outliers. So let's 'break' or 'cut-out' the y-axis
# into two portions - use the top (ax1) for the outliers, and the bottom
# (ax2) for the details of the majority of our data
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
fig.subplots_adjust(hspace=0.05)  # adjust space between axes

# plot the same data on both axes
ax1.plot(pts)
ax2.plot(pts)

# zoom-in / limit the view to different portions of the data
ax1.set_ylim(.78, 1.)  # outliers only
ax2.set_ylim(0, .22)  # most of the data

# hide the spines between ax and ax2
ax1.spines['bottom'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
#ax1.xaxis.tick_top()
#ax1.tick_params(labeltop=False)  # don't put tick labels at the top
#ax2.xaxis.tick_bottom()

# Now, let's turn towards the cut-out slanted lines.
# We create line objects in axes coordinates, in which (0,0), (0,1),
# (1,0), and (1,1) are the four corners of the axes.
# The slanted lines themselves are markers at those locations, such that the
# lines keep their angle and position, independent of the axes size or scale
# Finally, we need to disable clipping.

d = .5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
# ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
# ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

percentage = ['arPLS', 'asPLS', 'ResNet', 'UNet', 'ResUNet']
xIndex = np.arange(0, 10, 2)


RMSE = [59.75, 39.64, 6.11, 5.68, 5.40]
MAE =  [26.81, 21.27, 4.77, 4.43, 4.24]


bar_width = 0.4
bwidth = 0.3

d = 0.5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=50,
              linestyle="none", color='k', mec='k', mew=5, clip_on=False)

plt.figure(figsize=(45, 25))
plt.subplots_adjust(hspace=0.1)

plt.subplot(211)
ax = plt.gca()
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_visible(False)
ax.set_ylim([20, 60])
ax.plot([0, 0], [0, 0], transform=ax.transAxes, **kwargs)
ax.yaxis.set_tick_params(pad=18)

plt.bar(xIndex, height = RMSE, width=bwidth, label='RMSE', color='sandybrown', zorder=2)
plt.bar(xIndex+bar_width, height = MAE, width=bwidth, label='MAE', color='cornflowerblue', zorder=2)
plt.text(-0.7, 63, r'$\times 10^{-4}$', fontsize=60)
plt.yticks(range(20, 61, 10), fontsize=50)
plt.grid(axis='y', linewidth=3, linestyle='--')
plt.legend(fontsize=50)

plt.subplot(212)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_ylim([4, 6.5])
ax.plot([0, 0], [1, 1], transform=ax.transAxes, **kwargs)
ax.yaxis.set_tick_params(pad=18)

plt.bar(xIndex, RMSE, width=bwidth, color='sandybrown')
plt.bar(xIndex+bar_width, MAE, width=bwidth, color='cornflowerblue')

plt.xticks(xIndex+0.2, percentage, y=-0.02, fontsize=50)
plt.yticks(range(4, 7, 1), fontsize=50)

plt.grid(axis='y', linewidth=3, linestyle='--')

eps_fig  = plt.gcf()
eps_fig.savefig('C:\\Users\\CHENTIEJUN\\Desktop\\baseline\\chartBar.eps',format='eps',dpi=1000)


plt.show()
```
