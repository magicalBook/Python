```python
#Face detection use MTCNN in image

from mtcnn.mtcnn import MTCNN
import cv2
import os


detector = MTCNN()

Path = 'C:\\Users\\CHEN TIEJUN\\Desktop\\testDatas'
imageName = '1'

inputImgPath = os.path.join(Path,imageName)+'.jpg'
outputImgPath = Path+'\\result_'+imageName+'.jpg'

img = cv2.imread(inputImgPath)

face = detector.detect_faces(img)
#face = face[0]
for i in range(len(face)):
    box = face[i]['box']
    keypoints = face[i]['keypoints']

    cv2.rectangle(img,(box[0],box[1]),(box[0]+box[2],box[1]+box[3]),(255,0,0),2)
    cv2.circle(img,keypoints['left_eye'],1,(255,0,0),4)
    cv2.circle(img,keypoints['right_eye'],1,(255,0,0),4)
    cv2.circle(img,keypoints['nose'],1,(255,0,0),4)
    cv2.circle(img,keypoints['mouth_left'],1,(255,0,0),4)
    cv2.circle(img,keypoints['mouth_right'],1,(255,0,0),4)

cv2.imwrite(outputImgPath,img)
cv2.imshow('result',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
