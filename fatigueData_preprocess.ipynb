```python
import os
import shutil
import xlrd
import time

basePath = 'D:\\'
trainFolder = os.path.join(basePath, 'fatigueData')
imgFolder = 'depthCam_Driver_2_17_16_10_46'
imgPath = os.path.join(basePath, imgFolder)
excelPath = os.path.join(basePath, imgFolder) + '.xlsx'

excelData = xlrd.open_workbook(excelPath)
table = excelData.sheet_by_name('Sheet1')

beginTime = time.time()
for i,fatigueDegree in enumerate(table.col_values(1)[1:]):
    if fatigueDegree == 0:
        FDfolder = os.path.join(trainFolder,str(0))
        imgName = '{:0>10}'.format(i+1)+'.jpg'
        outimgName = 'b{:0>10}'.format(i+1)+'.jpg'
        outPath = os.path.join(FDfolder,outimgName)
        shutil.copy(os.path.join(imgPath,imgName),outPath)
        
    elif fatigueDegree == 1:
        FDfolder = os.path.join(trainFolder,str(1))
        imgName = '{:0>10}'.format(i+1)+'.jpg'
        outimgName = 'b{:0>10}'.format(i+1)+'.jpg'
        outPath = os.path.join(FDfolder,outimgName)
        shutil.copy(os.path.join(imgPath,imgName),outPath)
        
    elif fatigueDegree ==2:
        FDfolder = os.path.join(trainFolder,str(2))
        imgName = '{:0>10}'.format(i+1)+'.jpg'
        outimgName = 'b{:0>10}'.format(i+1)+'.jpg'
        outPath = os.path.join(FDfolder,outimgName)
        shutil.copy(os.path.join(imgPath,imgName),outPath)
        
endTime = time.time()

print(endTime-beginTime)
```

```python
from mtcnn.mtcnn import MTCNN
import cv2
import os

detector = MTCNN()

basePath = 'D:\\fatigueData'
for folderName in os.listdir(basePath):
    folderPath = os.path.join(basePath, folderName)
    for imgName in os.listdir(folderPath):
        img = cv2.cvtColor(cv2.imread(os.path.join(folderPath, imgName)), cv2.COLOR_BGR2RGB)
        
        try:
            face = detector.detect_faces(img)

            print('face: ', face)
            box = face[0]['box']

            cropped = img[box[1]:box[1]+box[3], box[0]:box[0]+box[2]]

            cropped = cv2.cvtColor(cropped, cv2.COLOR_RGB2BGR)

            cv2.imwrite(os.path.join(folderPath, imgName), cropped)
        except:
            continue
```
